# **CASES AND PROPERTIES**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Property_stolen_and_recovered.csv')

area_name_list = data['area_name'].unique().tolist()
area_name_list_index = np.arange(1, len(area_name_list) + 1, 1)
area_name_text_to_num = {i: j for i, j in zip(area_name_list, area_name_list_index)}
area_name_num_to_text = {j: i for i, j in zip(area_name_list, area_name_list_index)}

group_name_list = data['group_name'].unique().tolist()
group_name_list_index = np.arange(1, len(group_name_list) + 1, 1)
group_name_text_to_num = {i: j for i, j in zip(group_name_list, group_name_list_index)}
group_name_num_to_text = {j: i for i, j in zip(group_name_list, group_name_list_index)}

sub_group_name_list = data['sub_group_name'].unique().tolist()
sub_group_name_list_index = np.arange(1, len(sub_group_name_list) + 1, 1)
sub_group_name_text_to_num = {i: j for i, j in zip(sub_group_name_list, sub_group_name_list_index)}
sub_group_name_num_to_text = {j: i for i, j in zip(sub_group_name_list, sub_group_name_list_index)}

data['area_name'] = data['area_name'].map(area_name_text_to_num)
data['group_name'] = data['group_name'].map(group_name_text_to_num)
data['sub_group_name'] = data['sub_group_name'].map(sub_group_name_text_to_num)

data_sorted = data.sort_values(by=['area_name', 'group_name', 'sub_group_name', 'year'])

data_sorted['cases_recovery_rate'] = data_sorted['cases_property_recovered'] / data['cases_property_stolen']
data_sorted.fillna(1, inplace=True)

data_sorted['property_value_recovery_rate'] = data_sorted['value_of_property_recovered'] / data[
    'value_of_property_stolen']
data_sorted.fillna(1, inplace=True)
```

```python
data_sorted
```

```python
from sklearn.cluster import KMeans

# Selecting features for clustering
X = data_sorted[['cases_property_stolen', 'cases_property_recovered']]

# Initializing KMeans with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fitting KMeans model
kmeans.fit(X)

# Adding cluster labels to the dataframe
data_sorted['cluster'] = kmeans.labels_

# Visualizing clusters
plt.figure(figsize=(10, 6))

# Scatter plot for each cluster
for cluster_label in range(3):
    cluster_data = data_sorted[data_sorted['cluster'] == cluster_label]
    plt.scatter(cluster_data['cases_property_stolen'], cluster_data['cases_property_recovered'], label=f'Cluster {cluster_label+1}')

plt.title('Clustering of Property Stolen and Recovered Data')
plt.xlabel('Property Stolen')
plt.ylabel('Property Recovered')
plt.legend()
plt.grid(True)
plt.savefig('cluster_1.png',bbox_inches='tight')

plt.show()
```

```python
from sklearn.cluster import KMeans

# Selecting features for clustering
X = data_sorted[['value_of_property_stolen', 'value_of_property_recovered']]

# Initializing KMeans with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fitting KMeans model
kmeans.fit(X)

# Adding cluster labels to the dataframe
data_sorted['cluster'] = kmeans.labels_

# Visualizing clusters
plt.figure(figsize=(10, 6))

# Scatter plot for each cluster
for cluster_label in range(3):
    cluster_data = data_sorted[data_sorted['cluster'] == cluster_label]
    plt.scatter(cluster_data['value_of_property_stolen'], cluster_data['value_of_property_recovered'], label=f'Cluster {cluster_label+1}')

plt.title('Clustering of Value of Property Stolen and Recovered Data')
plt.xlabel('Value of Property Stolen')
plt.ylabel('Value of Property Recovered')
plt.legend()
plt.grid(True)
plt.savefig('cluster_2.png',bbox_inches='tight')

plt.show()
```

```python
def classify_cluster(cluster_label):
    if cluster_label == 0:
        return 'Low'
    elif cluster_label == 1:
        return 'Medium'
    elif cluster_label == 2:
        return 'High'

# Create new columns for case_recovery_class and property_recovery_class
data_sorted['case_recovery_class'] = data_sorted['cluster'].apply(classify_cluster)
data_sorted['property_recovery_class'] = data_sorted['cluster'].apply(classify_cluster)

data_sorted
```

```python
area_name = int(input("enter area name : {}".format(area_name_text_to_num)))
print('')
group_name = int(input("enter crime group : {}".format(group_name_text_to_num)))
print('')
data_filter = (data_sorted['area_name'] == area_name) & (data_sorted['group_name'] == group_name)
filtered_data = data_sorted[data_filter]

print("MODEL DIAGNOSIS")
print(f"The case recovery rate for {area_name_num_to_text[area_name]} and {group_name_num_to_text[group_name]} is: {filtered_data['case_recovery_class'].mode().iloc[0]}")
print(f"The property recovery rate for {area_name_num_to_text[area_name]} and {group_name_num_to_text[group_name]} is: {filtered_data['property_recovery_class'].mode().iloc[0]}")
```

# **COMPLAINTS AGAINST POLICE**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df = pd.read_csv('Complaints_against_police.csv')
selected_columns = ['Area_Name', 'Year', 'CPA_-_Cases_Registered',
                    'CPA_-_Complaints_Received/Alleged',
                    'CPB_-_Police_Personnel_Convicted',
                    'CPC_-_Police_Personnel_Dismissal/Removal_from_Service',
                    'CPC_-_Police_Personnel_Major_Punishment_awarded']

df = df[selected_columns]
# Rename columns
df.columns = ['area_name', 'year',
              'cases_registered', 'complaints_received',
              'police_convicted', 'police_removed_from_service', 'police_punished']

df['total_punishment_cases'] = df['police_convicted'] + df['police_removed_from_service'] + df[
    'police_punished']

# Calculate ratios
df['received_registered_ratio'] = df['cases_registered'] / df['complaints_received']
df['registered_police_punished_ratio'] = df['police_punished'] / df['cases_registered']

area_name_list = df['area_name'].unique().tolist()
area_name_list_index = np.arange(1, len(area_name_list) + 1, 1)
area_name_text_to_num = {i: j for i, j in zip(area_name_list, area_name_list_index)}
area_name_num_to_text = {j: i for i, j in zip(area_name_list, area_name_list_index)}

# df['area_name'] = df['area_name'].map(area_name_text_to_num)
df_sorted = df.sort_values(by=['area_name', 'year'])
# complaint_against_police_area = st.sidebar.selectbox('choose a state',
#                                                      options=area_name_text_to_num)
```

```python
from sklearn.cluster import KMeans

# Selecting features for clustering
X = df_sorted[['cases_registered', 'complaints_received']]

# Initializing KMeans with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fitting KMeans model
kmeans.fit(X)

# Adding cluster labels to the dataframe
df_sorted['cluster'] = kmeans.labels_

# Visualizing clusters
plt.figure(figsize=(10, 6))

# Scatter plot for each cluster
for cluster_label in range(3):
    cluster_data = df_sorted[df_sorted['cluster'] == cluster_label]
    plt.scatter(cluster_data['complaints_received'], cluster_data['cases_registered'], label=f'Cluster {cluster_label+1}')

plt.title('Clustering of cases registered and complaints received')
plt.xlabel('Complaints Received')
plt.ylabel('Cases Registered')
plt.legend()
plt.grid(True)
plt.savefig('cluster_3.png',bbox_inches='tight')

plt.show()

def classify_cluster(cluster_label):
    if cluster_label == 0:
        return 'Medium'
    elif cluster_label == 1:
        return 'High'
    elif cluster_label == 2:
        return 'Low'

# Create new columns for case_recovery_class and property_recovery_class
df_sorted['complaint_registration_class'] = df_sorted['cluster'].apply(classify_cluster)
# df_sorted
```
